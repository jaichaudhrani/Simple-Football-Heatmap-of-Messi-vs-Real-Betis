```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import mplsoccer
```

```python
df=pd.read_csv(r'C:\Users\jai\Desktop\messibetis.csv')
```

```python
df.head()
```

```python
customcmap=matplotlib.colors.LinearSegmentedColormap.from_list('custom cmap',['blue','yellow','red'])
```

```python
pitch=mplsoccer.VerticalPitch(pitch_type='opta',pitch_color='green',line_color='white',line_zorder=2)
fig,ax=pitch.draw(figsize=(3,6))
fig.set_facecolor('black')
pitch.kdeplot(df['x'],df['y'],ax=ax,cmap=customcmap,fill=True,n_levels=100,zorder=1)

fig.suptitle(y=0.90,t='Lionel Messi HeatMap',color='white',fontsize=40)
plt.show()
```

```python
pitch=mplsoccer.VerticalPitch(pitch_type='opta',pitch_color='green',line_color='white',line_zorder=2)
fig,ax=pitch.draw(figsize=(3,6))
fig.set_facecolor('black')
pitch.kdeplot(df['endX'],df['endY'],ax=ax,cmap=customcmap,fill=True,n_levels=100,zorder=1)

fig.suptitle(y=0.90,t='Lionel Messi HeatMap at the end',color='white',fontsize=40)
plt.show()
```
